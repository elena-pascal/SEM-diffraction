```python
par = 41.78214

def fX(s, Z, a1, a2, a3, a4, b1, b2, b3, b4):

    return Z - (par * s*s*(a1 * np.exp(-b1*s*s) +
                           a2 * np.exp(-b2*s*s) +
                           a3 * np.exp(-b3*s*s) +
                           a4 * np.exp(-b4*s*s)) )
```

```python
def fX_N(s):

    Z = 7
    a1 = 0.572
    b1 = 28.847
    
    a2 = 1.043
    b2 = 9.054
    
    a3 = 0.465
    b3 = 2.421
    
    a4 = 0.131
    b4 = 0.317

    return fX(s, Z, a1, a2, a3, a4, b1, b2, b3, b4)




def fX_Al(s):

    Z = 13
    a1 = 2.276
    b1 = 72.322
    
    a2 = 2.428
    b2 = 19.773
    
    a3 = 0.858
    b3 = 3.080
    
    a4 = 0.317
    b4 = 0.408

    return fX(s, Z, a1, a2, a3, a4, b1, b2, b3, b4)





def fX_Ga(s):

    Z = 31
    a1 = 2.321
    b1 = 65.602
    
    a2 = 2.486
    b2 = 15.458
    
    a3 = 1.688
    b3 = 2.581
    
    a4 = 0.599
    b4 = 0.351

    return fX(s, Z, a1, a2, a3, a4, b1, b2, b3, b4)



def fX_In(s):

    Z = 49
    a1 = 3.153
    b1 = 66.649
    
    a2 = 3.557
    b2 = 14.449
    
    a3 = 2.818
    b3 = 2.976
    
    a4 = 0.889
    b4 = 0.435

    return fX(s, Z, a1, a2, a3, a4, b1, b2, b3, b4)
```

#### lattice parameters

```python
#GaN wurtzite

a_GaN = 3.19 #A
c_GaN = 5.19 #A

#AlN wurtzite

a_AlN = 3.11
c_AlN = 4.98

#InN wurtzite

a_InN = 3.53
c_InN = 5.70
```

```python
def s_hkl(a, c, h, k, l):
    '''
    half the scattering angle divided by the wavelenght
    '''
    return ( (4/(3*a*a)) * (h*h + k*k + h*k) + (l*l/(c*c)))**(0.5)  /2.
    

# m-plane
print '    s m-plane:'

AlN_100 = s_hkl(a_AlN, c_AlN, 1,0,0)
print 's AlN in A^{-1}', AlN_100

GaN_100 = s_hkl(a_GaN, c_GaN, 1,0,0)
print 's GaN in A^{-1}', GaN_100

InN_100 = s_hkl(a_InN, c_InN, 1,0,0)
print 's InN in A^{-1}', InN_100

print

# c-plane
print '    s c-plane:'

AlN_001 = s_hkl(a_AlN, c_AlN, 0,0,1)
print 's AlN in A^{-1}', AlN_001

GaN_001 = s_hkl(a_GaN, c_GaN, 0,0,1)
print 's GaN in A^{-1}', GaN_001

InN_001 = s_hkl(a_InN, c_InN, 0,0,1)
print 's InN in A^{-1}', InN_001

print


# a-plane
print '    s a-plane:'

AlN_110 = s_hkl(a_AlN, c_AlN, 1,1,0)
print 'd AlN in A^{-1}', AlN_110

GaN_110 = s_hkl(a_GaN, c_GaN, 1,1,0)
print 's GaN in A^{-1}', GaN_110

InN_110 = s_hkl(a_InN, c_InN, 1,1,0)
print 's InN in A^{-1}', InN_110

print


# r-plane
print '  s  r-plane:'

AlN_112 = s_hkl(a_AlN, c_AlN, 1,-1,2)
print 's for AlN in A^{-1}', AlN_112

GaN_112 = s_hkl(a_GaN, c_GaN, 1,-1,2)
print 's for GaN in A^{-1}', GaN_112

InN_112 = s_hkl(a_InN, c_InN, 1,-1,2)
print 's for InN in A^{-1}', InN_112

print s_hkl(a_InN, c_InN, 1,1,1)
```

### Electron scattering factor: Mott-Bethe formula

```python
par_MB = 47.878 # V*A

def fe(s, Z, a1, a2, a3, a4, b1, b2, b3, b4):

    return par_MB*(a1 * np.exp(-b1*s*s) +
                   a2 * np.exp(-b2*s*s) +
                   a3 * np.exp(-b3*s*s) +
                   a4 * np.exp(-b4*s*s))
```

### Corrected Electron scattering factor using the DW factor at 300 K

```python
#calculated DW factors for individual elements at 300 K in A^2
DW_Al = 0.83 # A^2
DW_Ga = 0.49 # A^2
DW_In = 0.43 # A^2
DW_N_AlN = 0.28 # A^2
DW_N_GaN = 0.29 # A^2
DW_N_InN = 0.51 # A^2



# correction factor exponent: e^(-B*s^2)
def CF(DW_elm, s_elm):
    return np.exp(-DW_elm * s_elm * s_elm)



#corrected scattering factor: f_DWcor = f * CF
def fe_Ga_DWcor(s):
    return fe_Ga(s)*CF(DW_Ga, s)

def fe_Al_DWcor(s):
    return fe_Al(s)*CF(DW_Al, s)

def fe_In_DWcor(s):
    return fe_In(s)*CF(DW_In, s)

# N has different DW factors depending of the compund is part of
def fe_N_GaN_DWcor(s):
    return fe_N(s)*CF(DW_N_GaN, s)

def fe_N_AlN_DWcor(s):
    return fe_N(s)*CF(DW_N_AlN, s)

def fe_N_InN_DWcor(s):
    return fe_N(s)*CF(DW_N_InN, s)


```
